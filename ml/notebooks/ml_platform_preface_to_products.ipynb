# ML Platform: De Preface a Productos
**Versión**: 1.0  
**Objetivo**: Pipeline automatizado para extraer un prefacio PDF, normalizarlo, y generar 3 variantes de producto con modelo de precios

## 📋 Flujo Completo

```
1. Cargar & Parsear PDF
   └─ Extract metadata (title, author, ISBN, pages)
   └─ Extract Preface text (first N pages)
   └─ Segment into paragraphs

2. Detectar Parámetros Conceptuales
   └─ 8 entidades: Probabilistic ML, Transparency, Uncertainty Quantification, etc.
   └─ Keyword matching contra texto extraído
   └─ Asignar relevancia (alta/media/baja)

3. Normalizar a content.json
   └─ Schema: product_id, title (es/en), summary (es/en), parameters, metadata
   └─ Bilingual consistency checks
   └─ Source references (origin_phrase, relevance)

4. Definir 3 Variantes de Producto
   ├─ Ebook: PDF/EPUB ES/EN + glossary
   ├─ Course: 5 capítulos + notebooks + ejercicios + quizzes  
   └─ Mentoring: 10 sesiones 1:1 + simulaciones + docs regulatorios

5. Modelo de Precios con Activación de Parámetros
   ├─ Tier 0 (Intro): $0-50 → 2 parámetros → Ebook only
   ├─ Tier 1 (Professional): $50-200 → 4 parámetros (+2) → Ebook + Course
   ├─ Tier 2 (Premium): $200-600 → 6 parámetros (+2) → Ebook + Course + 5 mentoring
   └─ Tier 3 (Enterprise): $600-5000 → 8 parámetros (+2) → Full bundle + custom

6. Exportar Artefactos
   ├─ JSON: content.json, product_variants_*.json, pricing_tiers.json
   ├─ CSV: pricing_features.csv (con activación de features por tier)
   ├─ Markdown: Sales pages ES/EN para cada variante
   └─ Summary: export_summary.json + project_requirements.json

7. QA & CI/CD
   └─ Schema validation, bilingual checks, pricing rule enforcement
   └─ Notebook execution tests, translation review process
```

## 🎯 Validaciones Implementadas

- ✅ Regla: Cada tier activa ≥2 parámetros nuevos
- ✅ Bilingual consistency (all content ES/EN duplicated)
- ✅ Price ranges non-overlapping and monotonically increasing
- ✅ CSV export matches JSON definitions
- ✅ All product IDs unique and traceable

````python
def generate_product_page_markdown(product_id, variant_info, pricing_info, language="es"):
    """Genera plantilla markdown para página de producto"""
    
    title = variant_info["title"][language]
    description = variant_info["description"][language]
    price_min = pricing_info["price_range"]["min"]
    price_max = pricing_info["price_range"]["max"]
    
    lang_labels = {
        "es": {
            "features": "Características Incluidas",
            "price": "Precio",
            "learn_more": "Más Información",
            "get_access": "Obtener Acceso",
            "about": "Acerca de",
            "requirements": "Requisitos",
            "target": "Audiencia Objetivo",
            "resources": "Recursos Incluidos"
        },
        "en": {
            "features": "Included Features",
            "price": "Price",
            "learn_more": "Learn More",
            "get_access": "Get Access",
            "about": "About",
            "requirements": "Requirements",
            "target": "Target Audience",
            "resources": "Included Resources"
        }
    }
    
    labels = lang_labels[language]
    
    markdown = f"""# {title}

## {labels['about']}

{description}

## {labels['features']}

"""
    
    # Add features
    for feature in variant_info.get("deliverables", [])[:5]:
        if isinstance(feature, dict):
            markdown += f"- {feature.get('name', feature)}\n"
        else:
            markdown += f"- {feature}\n"
    
    markdown += f"""
## {labels['price']}

**${price_min:.2f} - ${price_max:.2f}**

[{labels['get_access']}](#cta)

## {labels['resources']}

- Interactive Content
- Community Access
- Regular Updates
- Email Support

---

*Last updated: {datetime.now().strftime('%Y-%m-%d')}*
"""
    
    return markdown

# Generar plantillas para cada variante y lenguaje
print("📝 Generando plantillas de productos...\n")

template_count = 0
for product_id, variant_info in PRODUCT_VARIANTS.items():
    # Obtener pricing info del primer tier que incluye este producto
    pricing_info = None
    for tier in PRICING_TIERS.values():
        if product_id in tier["products_included"]:
            pricing_info = tier
            break
    
    # Usar primer tier si no encuentra
    if not pricing_info:
        pricing_info = list(PRICING_TIERS.values())[0]
    
    for language in ["es", "en"]:
        markdown_content = generate_product_page_markdown(
            product_id, 
            variant_info, 
            pricing_info,
            language=language
        )
        
        filename = f"{product_id}_{language}.md"
        filepath = OUTPUT_DIR / "markdown" / filename
        
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(markdown_content)
        
        print(f"  ✓ {filename}")
        template_count += 1

print(f"\n✓ {template_count} plantillas markdown generadas")
````

````python
def extract_pdf_metadata(pdf_path: str) -> Dict[str, Any]:
    """Extrae metadatos básicos del PDF"""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            metadata = pdf.metadata
            num_pages = len(pdf.pages)
            
            return {
                "title": metadata.get("Title", "Unknown"),
                "author": metadata.get("Author", "Unknown"),
                "subject": metadata.get("Subject", "Unknown"),
                "creator": metadata.get("Creator", "Unknown"),
                "producer": metadata.get("Producer", "Unknown"),
                "creation_date": metadata.get("CreationDate", "Unknown"),
                "num_pages": num_pages,
                "isbn": None  # Extraer manualmente o via OCR
            }
    except Exception as e:
        print(f"❌ Error extrayendo metadata: {e}")
        return {}

def extract_preface_text(pdf_path: str, preface_end_page: int = 5) -> str:
    """Extrae el texto del Preface (primeras N páginas)"""
    preface_text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for i, page in enumerate(pdf.pages[:preface_end_page]):
                text = page.extract_text()
                if text:
                    preface_text += f"\n\n--- Página {i+1} ---\n{text}"
        return preface_text
    except Exception as e:
        print(f"❌ Error extrayendo Preface: {e}")
        return ""

def parse_preface_into_paragraphs(text: str) -> List[str]:
    """Segmenta el texto en párrafos"""
    # Split por dobles saltos de línea o por puntuación al final
    paragraphs = re.split(r'\n\n+', text.strip())
    return [p.strip() for p in paragraphs if p.strip() and len(p.strip()) > 20]

# Ejecutar extracción
print("📖 Extrayendo información del PDF...")

# Para propósitos de demostración, crearemos un Preface de ejemplo
# En producción, descomenta la siguiente línea y comenta el ejemplo
DEMO_PREFACE = """
Probabilistic Machine Learning offers a principled approach to understanding uncertainty in financial systems. 
This book bridges generative AI and classical probabilistic modeling, emphasizing transparency, explicability, 
and robustness. We focus on quantifying uncertainty through predictive intervals and calibrated densities, 
incorporating institutional knowledge while safeguarding against data-driven pitfalls. Our framework counters 
non-probabilistic model risks by enabling counterfactual simulations and regulatory documentation. Each concept—
from Bayesian inference to causal discovery—demonstrates how probabilistic thinking transforms finance and investing.
"""

# Si el PDF existe localmente, descomenta esto:
# pdf_metadata = extract_pdf_metadata(PDF_PATH)
# preface_raw = extract_preface_text(PDF_PATH, preface_end_page=3)

# Demostración con datos de ejemplo
pdf_metadata = {
    "title": "Probabilistic Machine Learning for Finance and Investing",
    "author": "Deepak K. Kanungo",
    "isbn": "978-XXXX-XXXXX",
    "num_pages": 250,
    "creation_date": "2024-01-01"
}
preface_raw = DEMO_PREFACE

# Parsear en párrafos
preface_paragraphs = parse_preface_into_paragraphs(preface_raw)

print(f"\n✓ PDF Metadatos:")
print(f"  Título: {pdf_metadata['title']}")
print(f"  Autor: {pdf_metadata['author']}")
print(f"  Páginas: {pdf_metadata['num_pages']}")
print(f"\n✓ Preface Paragraphs: {len(preface_paragraphs)} párrafos extraídos")
for i, para in enumerate(preface_paragraphs[:3]):
    print(f"  [{i+1}] {para[:100]}...")
````
